## Bar graph

```python
x = ['A', 'B', 'C']
y = [10, 15, 25]
# Create the bar graph
plt.bar(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 25]
# Create the bar graph
plt.barh(x, y)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')

# Display the graph
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 25]

plt.figure(figsize=(4,4)) ## Width: 5 inches, Height: 5 inches
# Create the bar graph
plt.bar(x, y,color=['r','g','y'])

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')
plt.grid()
# plt.legend()

# Display the graph
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 25]

plt.figure(figsize=(10,5))
# Create the bar graph
bars=plt.bar(x, y,color=['r','g','y'],label=['Category A', 'Category B', 'Category C'])

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')
# plt.grid()
# plt.legend()
plt.legend(bars, ['Category A', 'Category B', 'Category C'])
# Display the graph
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 25]

plt.figure(figsize=(10,2))
# Create the bar graph
plt.barh(x, y,color=['r','g','y'])

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Graph')
plt.grid()

# Display the graph
plt.show()
```

## scatter plot

## Histogram

```python
data = np.random.randn(1000)

# Create the histogram
plt.hist(data, bins=10)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the graph
plt.show()
```

## Pie Chart

```python

# Sample data
labels = ['A', 'B', 'C']
values = [35, 45, 25]
plt.figure(figsize=(5,5))
# Create the pie chart
plt.pie(values, labels=labels, autopct='%1.2f%%')
# plt.pie(values, labels=labels)

# Add title
plt.title('Pie Chart')
plt.legend(labels, loc='upper right') # upper left, lower left, lower right
# Display the graph
plt.show()
```

## Boxplot

```python
data = np.random.randn(100)

# Create the box plot
plt.boxplot(data)

# Add title
plt.title('Box Plot')

# Display the graph
plt.show()
```

## Area Plot

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the area plot
plt.fill_between(x, y)

# Add labels and title
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Area Plot')

# Display the graph
plt.show()
```
